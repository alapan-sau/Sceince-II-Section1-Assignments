```
import random
import matplotlib.pyplot as plt
import numpy as np
```

```
def func1(x):
    return 3 * (x ** 2)

def func2(x, y):
    return (x ** 2) * y
```

## Part 1

### $I_1 \; vs \; N $ 

```
def I1vN(N, a, b):
    
    I_val = []
    N_val = []

    n = 1
    sum = 0
    while n <= N:
        x = random.random() * (b-a)
        y = func1(x)

        sum += ( y * (b-a) )
        I_val.append(sum/n)
        N_val.append(n)
        n+=1

    return I_val, N_val
```

```
### I1 vs N

I1_val, N_val = I1vN(10000, 0, 1)

plt.figure(figsize=(15, 10))
plt.title("I1 vs N")
plt.xlabel("Number of Points")
plt.ylabel("Integration value")
plt.plot(N_val, I1_val)
plt.plot(N_val, np.ones(len(I1_val)))
plt.savefig("I1vN")
plt.show()
```

### $I_2 \; vs \; N $ 

```
def I2vN(N, a, b):

    I_val = []
    N_val = []

    n = 1
    sum = 0
    while n <= N:
        x = random.random() * (b-a)
        y = random.random() * (b-a)
        z = func2(x, y)

        sum += ( z * (b-a)**2 )
        I_val.append(sum/n)
        N_val.append(n)
        n+=1

    return I_val, N_val
```

```
### I2 vs N

I2_val, N_val = I2vN(10000, 0, 1)

plt.figure(figsize=(15, 10))
plt.title("I2 vs N")
plt.xlabel("Number of Points")
plt.ylabel("Integration value")
plt.plot(N_val, I2_val)
plt.plot(N_val, (1/6) * np.ones(len(I2_val)))
plt.savefig("I2vN")
plt.show()
```

## Part 2

### $I_1 \; vs \; T $ 

```
def I1vT(T, N, a, b):

    T_val = []
    I_val = []

    t = 1
    sum_over_trial = 0

    while t <= T:

        n = 1
        sum_over_point = 0
        while n <= N:
            x = random.random() * (b-a)
            y = func1(x)

            sum_over_point += ( y * (b-a) )
            n+=1
        
        avg_over_point = sum_over_point/ N

        sum_over_trial += avg_over_point

        I_val.append(sum_over_trial / t)
        T_val.append(t)
        t+=1
    
    return I_val, T_val
```

#### CASE1 :  $N = 20$

```
### I1 vs T(N = 20)

I1_val, T_val = I1vT(100, 20 , 0, 1)

plt.figure(figsize=(15, 10))
plt.title("I1 vs T (N = 20)")
plt.xlabel("Number of Trials")
plt.ylabel("Integration value")
plt.plot(T_val, I1_val)
plt.plot(T_val, 1 * np.ones(len(I1_val)))
plt.savefig("I1vT_N_20")
plt.show()

print ("Standard Deviation: " + str(np.std(np.array(I1_val))))
```

#### CASE2 : $ N = 1000$

```
### I1 vs T(N = 1000)

I1_val, T_val = I1vT(100, 1000 , 0, 1)

plt.figure(figsize=(15, 10))
plt.title("I1 vs T (N = 1000)")
plt.xlabel("Number of Trials")
plt.ylabel("Integration value")
plt.plot(T_val, I1_val)
plt.plot(T_val, 1 * np.ones(len(I1_val)))
plt.savefig("I1vT_N_1000")
plt.show()


print ("Standard Deviation: " + str(np.std(np.array(I1_val))))
```

## $ \sigma \; vs \; N (T=1000, I_1)$

```
def Std1vN(N, T, a, b):

    I_val = []
    N_val = []
    Std_val = []

    n = 1
    sum_over_point = 0

    while n <= N:
        t = 1
        sum_over_trial = 0
        while t <= T:
            x = random.random() * (b-a)
            y = func1(x)

            sum_over_trial += ( y * (b-a) )
            t+=1
        avg_over_trial = sum_over_trial/ T

        sum_over_point += avg_over_trial

        I_val.append(sum_over_point / n)
        Std_val.append(np.std(np.array(I_val)))
        N_val.append(n)
        n+=1
        
    return Std_val, N_val
```

```
### Std1 vs N(T = 1000)
Std1_val, N_val = Std1vN(500, 1000 , 0, 1)
plt.figure(figsize=(15, 10))
plt.title("Standard Deviation vs N")
plt.xlabel("Number of Points")
plt.ylabel("Standard Deviation")
plt.plot(N_val, Std1_val)
plt.savefig("Std1vT")
plt.show()
```
