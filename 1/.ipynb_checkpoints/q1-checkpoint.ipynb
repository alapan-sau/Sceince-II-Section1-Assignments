### A. Two Drunks meeting after $ n $ steps of random walk.
#### Analytical Approach:

We know a random walk of $ n $ steps by any of the drunk guys is a random sequence of $ L's $ and $ R's $. <br>
For e.g, $ LLRRLRRL... $ is such a sequence.

Total number of ways to make exactly $ l $ left steps out of total $ n $ steps (and $ n-l $ right steps)  = $ \large{n \choose l} $ <br>
Now for both of the drunk guys to meet at same place after $ n $ steps, they both need to make exactly same number of steps towards left and right. <br>
Total number of ways for both of the drunk guys to make exactly $ l $ steps towards left = $ \large{n \choose l}^{2} $ <br>
Total number of ways for both drunk guys to meet together = $\large{\sum_{l=0}^{n}{n\choose l}^2}$ = $ {2n \choose l} $

Total number of possible random walks for each guy = $ 2^n $ <br>
Total number of possible random walks for both guy together = $ 2^n * 2^n $ = $ 4^n $ <br>

Probabilty that both the drunk guys meet after walking $ N $ random steps =  $\large{\frac{{2n \choose l}}{4^n}}$ $ (ans)$


#### Computational Approach:

```python
import operator as op
from functools import reduce
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.special

## Random Walk Simulator
## return number of left steps in a new random walk
def randomPathSimul(iter):
    count = 0
    for i in range(iter):
        if random.uniform(0,1) < 0.5:
            count = count +1
    return count
```

```python
## python code to emulate the probabilistic simulation of meeting of two random walks
x = np.array(range(500))
y =[]
for iter in x:
    deno = 0
    nume = 0
    for sim_iter in range(1000):
        deno=deno+1
        num1 = randomPathSimul(iter)
        num2 = randomPathSimul(iter)

        if num1 == num2:
            nume=nume+1
    prob = float(nume)/float(deno)
    y.append(prob)




## python code to get the analytic graph for the above

x1 = np.array(range(500))
y1 =[]
for iter in x1:
    y1.append(0.25**iter * scipy.special.comb(2*iter,iter))
```

```python
plt.figure(figsize=(15,12))
plt.plot(x,y,'r',label="Computational")
plt.plot(x1,y1,'b',label="Analytical")
plt.legend()
plt.savefig('p_meeting.png')
```

### B. Random walk ending at origin after $ n $ steps.

#### Analytical Approach:

For, the random walk to end at centre, the walk should consist of equal number of left and right steps.<br>

Thus if $ n $ is $ even $, the number of such possible walks = $ \large{n \choose n/2} $ <br>
With an $ odd $ $ n $ , the number of such walks = $ 0 $ <br>
Total number of random walks of $ n $ steps = $ 2^n $ <br>



Thus, if $ n $ is  $ odd $ the probabilty of random walk to end at centre = $ 0 $ <br>
If $ n $ is  $ even $ the probabilty of random walk to end at centre = $\large{\frac{{n \choose n/2}}{2^n}}$ $ (ans)$



#### Computational Approach:

```python
## python code to emulate the probabilistic simulation of ending the random walk at origin
x2 = np.array(range(200))
y2=[]

for iter in x2:
    if iter % 2:
        y2.append(0)
    else:
        deno = 0
        nume = 0
        for sim_iter in range(1000):
            deno=deno+1
            if(randomPathSimul(iter) == iter/2):
                nume = nume+1
        y2.append(nume/deno)

## python code to plot analytic graph of the same
x3 = np.array(range(200))
y3 = []

for iter in x3:
    if iter % 2:
        y3.append(0)
    else:
        y3.append((0.5**iter * scipy.special.comb(iter,iter/2)))
```

```python
plt.figure(figsize=(15,12))
plt.plot(x2,y2,'r',label="Computational")
plt.plot(x3,y3,'b',label="Analytical", alpha=0.6)
plt.legend()
plt.savefig('p_origin.png')
```

### C. Mean displacement of a Random Walk.

#### Analytical Approach:

Let's define,
$ X_i $ be the random variable denoting the $ ith $ step out of $ n $ steps. <br>
Thus, $ X_i $ = $ -1 $ if the $ ith $ step is toward $ left $ and $ X_i $ = $ 1 $ if $ ith $ step is toward $ right $.

Therefore final displacement after $ n $ steps, $ d $ = ${\sum_{i=0}^{n} {X_i}} $. <br>

$ \implies{} E[d] = {\sum_{i=0}^{n} E[{X_i}]} $ <br><br>
Now, $ E[X_i]  = \frac{{1}}{2} * (1) + \frac{{1}}{2} * (-1)  = 0 $ <br>


$ \implies{} E[d] = n*0 = 0 $ $(ans)$<br><br>



#### Computational Approach:

```python
## Python code for simulating mean displacement

x4 = np.array(range(200))
y4=[]

for iter in x4:
    deno = 0
    nume = 0
    for sim_iter in range(10000):
        deno=deno+1
        nume = nume + iter - (2 * randomPathSimul(iter))
    y4.append(nume/deno)


## analytic code for the above
x5 = np.array(range(200))
y5=[]
for iter in x5:
    y5.append(0)
```

```python
plt.figure(figsize=(15,3))
plt.plot(x4,y4,'r',label="Computational")
plt.plot(x5,y5,'b',label="Analytical", alpha=0.6)
plt.legend()
plt.savefig('mean_displacement.png')
```

### D. Mean squared displacement of a Random Walk.

#### Analytical Approach:

Let's define,
$ X_i $ be the random variable denoting the $ ith $ step out of $ n $ steps. <br>
Thus, $ X_i $ = $ -1 $ if the $ ith $ step is toward $ left $ and $ X_i $ = $ 1 $ if $ ith $ step is toward $ right $.

Therefore final displacement after $ n $ steps, $ d $ = ${\sum_{i=0}^{n} {X_i}} $. <br>
$ \implies{} d^2 = (\sum_{i=0}^{n} {X_i} )^2$ <br>
$ \implies{} d^2 = (X_1^2 + X_2^2 + ... + X_n^2) + 2(X_1X_2 + X_1X_3 + ... + X_{n-1}X_{n})$ <br>
$ \implies{} d^2 = \sum_{i=0}^{n} {X_i}^2  + 2\sum_{i\neq{}j,i=0,j=0}^{n} {X_iX_j}$

$ \implies{} E[d^2] = \sum_{i=0}^{n} E[{X_i}^2]  + 2\sum_{i\neq{}j,i=0,j=0}^{n} E[{X_iX_j}]$


Now, $ E[X_i^2]  = \frac{{1}}{2} * (1)^2 + \frac{{1}}{2} * (-1)^2  = 1$ <br>
Similarly, $ E[X_iX_j]  =  \frac{{1}}{4} * (1* 1) + \frac{{1}}{4} * (-1* 1) + \frac{{1}}{4} * (1* -1) + \frac{{1}}{4} * (-1* -1) = 0$

$ \implies{} E[d^2] = n*1 + 0 = n$.

Thus the mean-squared displacement after $ n $ steps = $ n $ $ (ans) $

#### Computational Approach:

```python
## Python code for simulating mean squared displacement
x6 = np.array(range(200))
y6=[]

for iter in x6:
    deno = 0
    nume = 0
    for sim_iter in range(10000):
        deno=deno+1
        nume = nume + (iter - (2 * randomPathSimul(iter)))**2
    y6.append(nume/deno)


## analytic values for the above
x7 = np.array(range(200))
y7=[]
for iter in x7:
    y7.append(iter)
```

```python
plt.figure(figsize=(15,12))
plt.plot(x6,y6,'r',label="Computational")
plt.plot(x7,y7,'b',label="Analytical", alpha=0.6)
plt.legend()
plt.savefig('meansq_displacement.png')
```
