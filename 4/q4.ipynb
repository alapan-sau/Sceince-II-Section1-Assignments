```
import matplotlib.pyplot as plt
import numpy as np
```

```
dx = 0.125
dt = 0.001
D = 4
L = 8
X = np.arange(-L, L+dx, dx)
N = [10, 100, 1000, 10000]

def one_dimensional_diffusion():
    plt.figure(figsize=(10, 10))
    plt.xlabel("Displacement")
    plt.ylabel("Probability")
    for n in N:
        probability = np.zeros( (n+1, len(X)) )

        # set initial probability
        probability[0][ int(len(X)/2) ] = 1
    
        for i in range(1,n+1):
            for j in range(len(X)-1):
                probability[i][j] = probability[i-1][j] + (D*dt/(dx**2)) * (probability[i-1][j+1] - 2*probability[i-1][j] + probability[i][j-1])
        plt.plot(X, probability[n] , label = f"N = {n}")
    plt.legend()
    plt.savefig("1D")
    plt.show()


one_dimensional_diffusion()
```

```
dx = 0.125
dy = 0.125
dt = 0.001
# D_x = 3
# D_y = 2
L = 4
X = np.arange(-L, L+dx, dx)
Y = np.arange(-L, L+dy, dy)
N = [10, 100, 1000, 10000]

def two_dimensional_diffusion(D_x, D_y):
    for n in N:
        probability = np.zeros( (n+1, len(X), len(Y)) )

        # set initial probability
        probability[0][ int(len(X)/2) ][int(len(Y)/2)] = 1
    
        for i in range(1,n+1):
            for j in range(len(X)-1):
                for k in range(len(Y)-1):
                        x_temp = D_x / (dx**2) * (probability[i-1][j+1][k] - 2*probability[i-1][j][k] + probability[i-1][j-1][k])
                        y_temp = D_y / (dy**2) * (probability[i-1][j][k+1] - 2*probability[i-1][j][k] + probability[i-1][j][k-1])
                        probability[i][j][k] =  probability[i-1][j][k] + dt * (x_temp + y_temp)

        plt.figure(figsize=(10,10))
        plt.title("2D Diffusion (N=%d, Dx=%f, Dy=%f)"%(n,D_x,D_y))
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.contour(X, Y, probability[n])
        plt.savefig("2D_N_%d_Dx_%f_Dy_%f.png"%(n,D_x,D_y))
        plt.show()

# two_dimensional_diffusion()
```

```
two_dimensional_diffusion(3,3)
```

```
two_dimensional_diffusion(3,2)
```

```
two_dimensional_diffusion(2,3)
```
