```
##importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```
N = 100
spin_values = np.array([1,-1])
initial_spin_values = np.random.choice(spin_values,size = N)

J = 1.0
iter = 500000 

temperatures = np.linspace(0.05,4,10)

E_avg = []
M_avg = []

for T in temperatures:

    beta = 1.0/T 
    spin_array = initial_spin_values

    e = 0
    for k in range(N-1):
        e += -J*spin_array[k]*spin_array[k+1]

    E = []
    M = []

    m = np.sum(spin_array)

    ## run all simulations    
    for i in range(iter):

        r_n = np.random.randint(0,N)
        sigma_mu = spin_array[r_n]

        delta_m = -2*sigma_mu    

        delta_e = 0
        if(r_n == 0):
            delta_e = 2*J*sigma_mu*(spin_array[1]+spin_array[N-1])
        elif(r_n == N-1):
            delta_e = 2*J*sigma_mu*(spin_array[N-2]+spin_array[0])
        else:
            delta_e = 2*J*sigma_mu*(spin_array[r_n+1]+spin_array[r_n-1])
    

        if( np.random.uniform() < np.exp( -beta*delta_e ) ):
            spin_array[r_n] = -1*spin_array[r_n]
            e += delta_e
            m += delta_m

        if(i>=100000 and i%N==0):  
            E.append(e)
            M.append(abs(m))

    E_avg.append(np.mean(np.array(E)))
    M_avg.append(np.mean(np.array(M)))
        
```

```
T_theory = np.linspace(0.01,5,1000)
E_theory = -N*J*np.tanh(J/T_theory)


#########################

plt.figure(figsize = (15, 10))
plt.title(" Average Energy vs temperature")
plt.scatter(temperatures,E_avg,color = 'k',label = "Computatinal")
plt.plot(T_theory,E_theory, label = "Theoretical")
plt.ylabel("Energy")
plt.xlabel("Temperature")
plt.legend()
plt.savefig('energy_temp')
plt.show()

plt.figure(figsize = (15, 10))
plt.title(" Magnetization M vs temperature")
plt.plot(temperatures, np.array(M_avg)/N, label = "Computatinal")
plt.ylabel("Magnetisation")
plt.xlabel("Temperature")
plt.legend()
plt.savefig('magnetisation_temp')
plt.show()
```
