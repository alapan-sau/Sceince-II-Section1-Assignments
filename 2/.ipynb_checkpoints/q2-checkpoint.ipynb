## Computational Approximation of $\pi $ :

```
import numpy as np
import random
import matplotlib.pyplot as plt

num_of_iter = np.array(range(1,10000))
appx_val = []
count = 0
for iter in num_of_iter:
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)

    if(x**2 + y**2) <= 1:
        count = count + 1
    
    appx_val.append(4*count/iter)
```

```
plt.figure(figsize=(15,12))
plt.plot(num_of_iter,appx_val,'r',label="Computational")
plt.axline((0,3.14),(10000,3.14))
plt.legend()
plt.savefig('PIvN.png')
```
