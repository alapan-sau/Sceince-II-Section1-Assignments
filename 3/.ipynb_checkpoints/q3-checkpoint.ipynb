# Question 3:

For a one dimensional Harmonic Oscillator, the Hamiltonian can be expressed as:
$$
H(x,p) = \frac{1}{2}kx^2 + \frac{p^2}{2m}
$$<br>

The Hamilotinian Equations are:
$$
\frac{dx}{dt}= \frac{\partial H}{\partial p} \;\;\;\;\;\;\;\;\;\;\;\;\;(1)
$$

$$
\frac{dp}{dt}= -\frac{\partial H}{\partial x} \;\;\;\;\;\;\;\;\;\;\;\;(2) 
$$



From (1),
$$
\frac{dx}{dt}= \frac{\partial H}{\partial p}  = \frac{p}{m}
$$
From (2),
$$
\frac{dp}{dt}= -\frac{\partial H}{\partial x} = -kx
$$


```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = [1, 2, 3, 4, 5]
p = [1, 2, 3, 4, 5]
k = 1/5
m = 2
dt = 0.001
X_t = np.zeros((len(x), 100000))
P_t = np.zeros((len(p), 100000))

X_t[0][0] = 0
P_t[0][0] = 2

for i in range(len(x)):
    X_t[i][0] = x[i]
    P_t[i][0] = p[i]

    for j in range(1,100000):

        dp = -k*X_t[i][j-1]*dt
        dx = (P_t[i][j-1]/m)*dt

        X_t[i][j] = X_t[i][j-1] + dx
        P_t[i][j] =  P_t[i][j-1] + dp

plt.figure(figsize=(12,10))
for i in range(len(x)):
    H = 1/2*(k*(x[i]**2)+(p[i]**2)/m)
    plt.plot(X_t[i], P_t[i], label = f"H={H}")

plt.legend()
plt.xlabel("Displacement")
plt.ylabel("Momentum")
plt.title("Displacement vs Momentum")
plt.savefig("Displacement_vs_Momentum.png")
plt.show()
```

We know Mean Squared Displacement for multiple-particle-system is:

$$
MSD(t) = \langle r^2(t) \rangle \; = \; \langle |r_i(t) - r_i(0)|^2 \rangle
$$

In our case, we have a single particle, so no average is required.

```
plt.figure(figsize=(15,10))
for i in range(len(x)):
    plt.plot((X_t[i]-x[i])**2)


plt.xlabel("Time")
plt.ylabel("Mean Square")
plt.title("Mean Square vs Time")
plt.savefig("Mean_Square_vs_Time")
plt.show()
```
